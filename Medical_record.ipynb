```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (8, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
medical_df.age.describe()
```

```python
medical_df.bmi.describe()
```

```python
fig = px.histogram(medical_df,
                   x='age',
                   marginal='box',
                   nbins=47,
                   title='Distribution of Age')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig = px.histogram(medical_df,
                   x='bmi',
                   marginal='box',
                   color_discrete_sequence=['green'],
                   title='Distribution of BMI (Body Mass Index)')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical_df,
                   x='charges',
                   marginal='box',
                   color='smoker',
                   color_discrete_sequence=['green', 'grey'],
                   title='Annual Medical Charges')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical_df,
                 x='charges',
                 marginal='box',
                 color='sex',
                 color_discrete_sequence=['Green','Orange'],
                 title='Anual medical Charges(sex)')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical_df,
                 x='charges',
                 marginal='box',
                 color='region',
                 color_discrete_sequence=['Blue','Black'],
                 title='Anual Medical Charges(Region)')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
medical_df.smoker.value_counts()
```

```python
fig = px.histogram(medical_df, x='smoker', color='sex', title='Smoker', marginal='box')

fig.update_layout(bargap=0.1)
fig.show()
```

```python
px.histogram(medical_df, x='smoker', color='children', title='Smoker')
```

```python
fig= px.scatter (medical_df,
                y='charges',
                x='age',
                color='smoker',
                hover_data=['sex'],
                opacity=0.8,
                title='Age vs Charges')
fig.update_traces(marker_size=5)
fig.show()
```

```python
fig = px.scatter(medical_df,
                 x='bmi',
                 y='charges',
                 color='smoker',
                 opacity=0.8,
                 hover_data=['sex'],
                 title='BMI vs. Charges')
fig.update_traces(marker_size=5)
fig.show()
```

```python
fig_children = px.violin(medical_df, y="charges", x="children", box=True, points="all")
fig_children.show()
```

```python
fig = px.violin(medical_df, y="charges", x="sex", box=True, points="all")
fig.show()
```

```python
fig_region = px.violin(medical_df, y="charges", x="region", box=True, points="all")
fig_region.show()
```

```python
smoker_values = {'no': 0, 'yes': 1}
smoker_numeric = medical_df.smoker.map(smoker_values)
smoker_numeric
```

```python
medical_df.corr(numeric_only=True)
```

```python
# Convert categorical variables to numeric values
medical_df['sex'] = medical_df['sex'].astype('category').cat.codes

# Calculate the correlation matrix
corr_matrix = medical_df.corr(numeric_only=True)

# Plot the heatmap
sns.heatmap(corr_matrix, cmap='Reds', annot=True)
plt.title('Correlation Matrix');
```
